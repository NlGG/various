```python
%matplotlib inline
import networkx as nx
import pylab as p
```

```python
class Node:
    def __init__(self, x):
        self.data  = x
        self.parent = None
        self.child = None

def insert(node, x):
    node.child = Node(x)
    return node
```

```python
node = Node(1)
node = insert(node, 3)
node.child.parent = node
node = node.child
node = insert(node, 10)
node.child.parent = node
node = node.child
```
